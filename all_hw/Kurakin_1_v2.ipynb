# Описание задачи

---

* id: уникальный идентификатор
* gender: "мужской", "женский" или "другой"
* age: возраст пациента
* hypertension: 0, если у пациента нет гипертонии, 1, если у пациента есть гипертония
* heart_disease: 0, если у пациента нет никаких сердечных заболеваний, 1, если у пациента есть сердечное заболевание
* ever_married: "Нет" или "Да"
* work_type: "дети", "гос_работа", "никогда_не_работал", "частный" или "самозанятый"
* Residence_type: "сельский" или "городской"
* avg_glucose_level: средний уровень глюкозы в крови
* bmi: индекс массы тела
* smoking_status: "ранее курил", "никогда не курил", "курит" или "неизвестно "*.
* stroke: 1, если у пациента был инсульт, или 0, если нет.
* Примечание: "Неизвестно" в статусе курения означает, что информация для данного пациента недоступна.


> Изначально, цель задачи классифицировать пациентов по признаку stroke - был ли у пациента инсульт в прошлом, однако в данном исследовании цель была измененена, в качестве эксперимента, целевой переменной стало heart_disease (наличие или отсутствие заболеваний сердца) по приведенным в данных параметрам.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('sample_data/heart_2020_cleaned.csv')
df.head()
```

```python
disease = df.groupby('HeartDisease').size()
disease
```

```python
X = df.drop('HeartDisease', axis=1)
y = df['HeartDisease']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=True)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

# Обучение моделей, подбор параметров

```python
from sklearn import metrics

#best_params_ = {'dtree__criterion': 'entropy', 'dtree__splitter': 'best', 'min_samples_split': 4}


dtree = DecisionTreeClassifier(criterion = 'entropy', splitter = 'best', min_samples_split = 4, random_state = 42)
dtree.fit(X_train, y_train)
dtree_predictions = dtree.predict(X_test)
print(classification_report(y_test, dtree_predictions))
metrics.plot_roc_curve(dtree, X_test, y_test)
```

```python
#best_params_ = {'rfc__criterion': 'entropy', 'rfc__n_estimators': 150, 'rfc__min_samples_split': 2}

n_estimators=150, 

rfc = RandomForestClassifier(criterion = 'entropy', n_estimators = 150, rfc_min_samples_split = 2)
rfc.fit(X_train, y_train)
rfc_prediction = rfc.predict(X_test)
metrics.plot_roc_curve(rfc, X_test, y_test)
print(classification_report(y_test, rfc_prediction))
```

```python
#best_params_ = {'knn__leaf_size': 15, 'knn__n_neighbors': 7, 'knn__weights': 'uniform'}

knn = KNeighborsClassifier(leaf_size = 15, n_neighbors =  7, weights= "uniform")
knn_model = knn.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
metrics.plot_roc_curve(knn_model, X_test, y_test)
print(classification_report(y_test, y_pred))
```
